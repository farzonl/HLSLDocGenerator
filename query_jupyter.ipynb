```python
from IPython.display import HTML, display
import pandas as pd
from query import gen_csv_data_base, run_dxc, query_dxil, md_table_to_dict, full_md_filepath, gen_remaining_opcode_data
```

```python
hlsl_to_dxil_op = run_dxc()
```
